```python
import wrds
```

```python
conn = wrds.Connection()
```

```python
conn.list_libraries().sort()
type(conn.list_libraries())
```

```python
conn.list_tables(library='crspa')
```

```python

# Extract first 5 obs from comp.company

company = conn.get_table(library='crsp_a_stock', table='dsf', obs=5)
company.columns
```

```python
company
```

```python
# Narrow down the specific columns to extract

company_narrow = conn.get_table(library='crsp', table='stocknames',
                                columns = ['ticker', 'comnam', 'shrcd'], obs=5)
company_narrow.shape
```

```python
company_narrow
```

```python
# Select one stock's monthly price
# from 2019 onwards

apple = conn.raw_sql("""select DISTINCT ON (date)
                        from crsp.dsf
                        where dsf.date BETWEEN '01/01/2019'  AND '10/10/2019'
                        group by date_part('wk',dsf.date)""",
                     date_cols=['date'])

apple
```

```python
# Select one stock's monthly price
# from 2019 onwards

apple = conn.raw_sql("""select date
                        from crsp.dsf
                        where dsf.date BETWEEN '01/01/2019'  AND '10/10/2019'
                        group by date_part('wk',date)""",
                     date_cols=['date'])

apple
```
